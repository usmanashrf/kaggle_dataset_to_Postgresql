```python
from sqlalchemy import create_engine
import pandas as pd

# Database credentials
DB_USER = 'your_neon_username'
DB_PASSWORD = 'your_neon_password'
DB_HOST = 'your_neon_host'
DB_PORT = '5432'  # Default PostgreSQL port
DB_NAME = 'your_neon_database'

# Create a connection string
DATABASE_URL = f'postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}'

# Create the database engine
engine = create_engine(DATABASE_URL)
```

```python
# Save dataset to PostgreSQL
TABLE_NAME = 'amazon_products'

products.to_sql(TABLE_NAME, con=engine, if_exists='replace', index=False)

print(f"Dataset successfully saved to the table '{TABLE_NAME}' in Neon DB!")
```

```python
# Save dataset to PostgreSQL
TABLE_NAME = 'amazon_categories'

categories.to_sql(TABLE_NAME, con=engine, if_exists='replace', index=False)

print(f"Dataset successfully saved to the table '{TABLE_NAME}' in Neon DB!")
```
